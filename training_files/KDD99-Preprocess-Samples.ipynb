## Clean and Generalize Datasets

```python
# Drop difficulty_level column from train and test sets
train_df.drop(['difficulty_level'], axis=1, inplace=True)
```

```python
# Function to generalize attack types in the datasets
def generalize_attack_types(df):
    df.label.replace(['normal'], "NORMAL", inplace=True)
    df.label.replace(['ipsweep','mscan','nmap','portsweep','saint','satan'],'PROBE',inplace=True)
    df.label.replace(['buffer_overflow','loadmodule','perl','ps','rootkit','sqlattack','xterm'],'U2R',inplace=True)
    df.label.replace(['apache2', 'back', 'land', 'neptune', 'mailbomb', 'pod', 'processtable', 'smurf', 'teardrop', 'udpstorm', 'worm'],'DOS',inplace=True)
    df.label.replace(['ftp_write', 'guess_passwd', 'httptunnel', 'imap', 'multihop', 'named', 'phf', 'sendmail', 'snmpgetattack', 'snmpguess', 'spy', 'warezclient', 'warezmaster', 'xlock', 'xsnoop'],'R2L',inplace=True)
```

```python
# Generalize attack types on train and test sets
generalize_attack_types(train_df)
```

```python
train_df['label'].value_counts()
```

```python
def generalize_attack_types_kdd99(df):
    df.label.replace(['normal.'], "NORMAL", inplace=True)
    df.label.replace(['ipsweep.','mscan.','nmap.','portsweep.','saint.','satan.'],'PROBE',inplace=True)
    df.label.replace(['buffer_overflow.','loadmodule.','perl.','ps.','rootkit.','sqlattack.','xterm.'],'U2R',inplace=True)
    df.label.replace(['apache2.', 'back.', 'land.', 'neptune.', 'mailbomb.', 'pod.', 'processtable.', 'smurf.', 'teardrop.', 'udpstorm.', 'worm.'],'DOS',inplace=True)
    df.label.replace(['ftp_write.', 'guess_passwd.', 'httptunnel.', 'imap.', 'multihop.', 'named.', 'phf.', 'sendmail.', 'snmpgetattack.', 'snmpguess.', 'spy.', 'warezclient.', 'warezmaster.', 'xlock.', 'xsnoop.'],'R2L',inplace=True)
```

## One-Hot Encoding

```python
def oh_encoder(df):
    # List of columns to be one-hot encoded
    columns_to_encode = ["protocol_type", "service", "flag"]

    # Apply one-hot encoding to each of the specified columns
    for column in columns_to_encode:
        one_hot_encoded_data = pd.get_dummies(df[column], prefix=column, dtype=int)
        df = pd.concat([df.drop(column, axis=1), one_hot_encoded_data], axis=1)
    return df
```

```python
train_df = oh_encoder(train_df)
kdd99 = oh_encoder(kdd99)
```

```python
# Save each label into a separate file.
for label in kdd99['label'].unique():
    # Filter the DataFrame based on the label
    df_filtered = kdd99[kdd99['label'] == label]

    # Check if the filtered DataFrame has at least 50 rows
    if len(df_filtered) >= 100:
        # Randomly sample 50 rows from the filtered DataFrame
        df_sampled = df_filtered.sample(n=100)
    else:
        df_sampled = df_filtered
    
    # Construct a file name based on the label
    file_name = f'{label}.csv'
    
    # Save the filtered DataFrame to a CSV file
    df_sampled.to_csv(file_name, index=False)

    print(f'Saved {len(df_sampled)} records to {file_name}')
```
