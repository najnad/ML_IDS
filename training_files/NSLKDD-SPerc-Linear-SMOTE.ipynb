## Clean and Generalize Datasets

```python
# Drop difficulty_level column from train and test sets
train_df.drop(['difficulty_level'], axis=1, inplace=True)
test_df.drop(['difficulty_level'], axis=1, inplace=True)
```

```python
# Function to generalize attack types in the datasets
def generalize_attack_types(df):
    df.label.replace(['normal'], "NORMAL", inplace=True)
    df.label.replace(['ipsweep','mscan','nmap','portsweep','saint','satan'],'PROBE',inplace=True)
    df.label.replace(['buffer_overflow','loadmodule','perl','ps','rootkit','sqlattack','xterm'],'U2R',inplace=True)
    df.label.replace(['apache2', 'back', 'land', 'neptune', 'mailbomb', 'pod', 'processtable', 'smurf', 'teardrop', 'udpstorm', 'worm'],'DOS',inplace=True)
    df.label.replace(['ftp_write', 'guess_passwd', 'httptunnel', 'imap', 'multihop', 'named', 'phf', 'sendmail', 'snmpgetattack', 'snmpguess', 'spy', 'warezclient', 'warezmaster', 'xlock', 'xsnoop'],'R2L',inplace=True)
```

```python
# Generalize attack types on train and test sets
generalize_attack_types(train_df)
generalize_attack_types(test_df)
```

## One-Hot Encoding

```python
def oh_encoder(df):
    # List of columns to be one-hot encoded
    columns_to_encode = ["protocol_type", "service", "flag"]

    # Apply one-hot encoding to each of the specified columns
    for column in columns_to_encode:
        one_hot_encoded_data = pd.get_dummies(df[column], prefix=column, dtype=int)
        df = pd.concat([df.drop(column, axis=1), one_hot_encoded_data], axis=1)
    return df
```

```python
train_df = oh_encoder(train_df)
```

```python
test_df = oh_encoder(test_df)
```

## Applying SMOTE to the Dataset

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(train_df, y_train_binary)
```

## Feature Selection: SelectPercentile

```python
from sklearn.feature_selection import SelectPercentile, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
best_sperc = 0
best_nest = 0
max_acc = 0
for j in range(1,10):
    for i in range(10,90):
        sperc = SelectPercentile(chi2, percentile=i).fit(X_train_smote, y_train_smote)
        X_selected = sperc.transform(X_train_smote)
        clf = RandomForestClassifier(n_estimators=j, n_jobs=-1, random_state=42)
        clf.fit(X_selected, y_train_smote)
        
        # Transform test data
        X_test_selected = sperc.transform(test_df)
    
        # Predict using resulting model
        y_pred = clf.predict(X_test_selected)
    
    
        # Calculate accuracy and print results
        accuracy = accuracy_score(y_test_binary, y_pred)
        if accuracy > max_acc:
                max_acc = accuracy
                best_nest = j
                best_sperc = i

print(f"Best S_PERC: {best_sperc} | Best NEST: {best_nest} | Accuracy: {max_acc}")
```

```python
sperc = SelectPercentile(chi2, percentile=71).fit(train_df, y_train_binary)
X_selected = sperc.transform(train_df)
clf = RandomForestClassifier(n_estimators=2, n_jobs=-1, random_state=42)
clf.fit(X_selected, y_train_binary)
# Transform test data
X_test_selected = sperc.transform(test_df)

# Predict using resulting model
y_pred = clf.predict(X_test_selected)

# Calculate accuracy and print results
accuracy = accuracy_score(y_test_binary, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test_binary, y_pred))
```
